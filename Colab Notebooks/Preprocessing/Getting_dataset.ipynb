```python
train_audio_path = r"/content/dataset/test/audio/"

all_wave = []
all_features = []

waves = [f for f in os.listdir(train_audio_path) if f.endswith('.wav')]

for wav in tqdm(waves):
    samples, sample_rate = librosa.load(path=train_audio_path + wav)
    if samples.shape[0] < 50000:
       samples = np.pad(samples, (0, 50000 - samples.shape[0]))
    else:
       samples = samples[0:50000]
    features = librosa.feature.mfcc(y=samples, sr=sample_rate)
    #all_wave.append(samples)
    all_features.append(features)

fnames = pd.DataFrame(waves, columns=["names"])
fnames.to_csv('/content/drive/MyDrive/Pro2/test_set/fnames_update050522.csv', index=False)
```
